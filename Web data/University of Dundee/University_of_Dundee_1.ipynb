```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import os

# Initialize Selenium WebDriver
driver = webdriver.Chrome()

# Create a directory to store the text files
os.makedirs("University_of_Dundee/", exist_ok=True)

# Iterate through each row in df1 (primary DataFrame)
for index, row in df.iterrows():
    degree_name = row['name']
    url1 = row['link']
    urls = [url1]  # Add the first DataFrame's URL

    # Add URLs from df2, df3, etc., based on the same index
    try:
        url2 = df1.loc[index, 'link']
        urls.append(url2)
    except KeyError:
        pass  # If the index does not exist in df2, skip it

    try:
        url3 = df2.loc[index, 'link']
        urls.append(url3)
    except KeyError:
        pass  # If the index does not exist in df3, skip it

    content = []  # List to store aggregated content

    try:
        for url in urls:
            # Open the URL
            driver.get(url)

            # Wait for the page content to load
            WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.CLASS_NAME, 'layout-container')))

            # Get page source after content is fully loaded
            soup = BeautifulSoup(driver.page_source, 'html.parser')

            # Find the div with specific id and class
            body_container = soup.find('div', {'class': 'layout-container'})

            if body_container:
                # Remove any header or footer elements inside the main div (if they exist)
                for tag in body_container.find_all(['header', 'footer']):
                    tag.decompose()  # Remove these tags and their content

                # Extract all text from the main content
                text_content = body_container.get_text(separator='\n', strip=True)
                content.append(f"Content from {url}:\n{text_content}\n\n")
            else:
                content.append(f"No main content found for URL: {url}\n")

        # Create a valid filename by replacing invalid characters
        filename = f"University_of_Dundee/{degree_name.replace('/', '_').replace(' ', '_')}.txt"

        # Save the aggregated content to a single file
        with open(filename, "w", encoding="utf-8") as file:
            # Write the content
            for section in content:
                file.write(section)

        print(f"Aggregated content successfully saved to '{filename}'.")

    except Exception as e:
        print(f"Failed to retrieve content for {degree_name}. Error: {e}")

# Close the Selenium driver
driver.quit()
```
