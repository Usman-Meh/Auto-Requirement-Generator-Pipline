```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Specify the URL of the webpage
url = 'https://www.mnsu.edu/academics/degree-programs-and-certificates/'  # Replace with your actual URL

# Step 2: Send a GET request to the webpage
response = requests.get(url)

# Step 3: Check if the request was successful
if response.status_code == 200:
    # Step 4: Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Step 5: Find the specific table with the class 'styled-table'
    table = soup.find('table', {'class': 'table styled-table'})

    # Step 6: Iterate through the rows <tr> inside <tbody>
    data = []
    if table:
        tbody = table.find('tbody')
        rows = tbody.find_all('tr')

        for row in rows:
            # Step 7: Iterate through each <td> within the row
            columns = row.find_all('td')
            for column in columns:
                # Step 8: Find all <li> elements within the <td>
                list_items = column.find_all('li')
                for item in list_items:
                    # Step 9: Extract <a> tag with program name and link
                    link_tag = item.find('a')
                    if link_tag:
                        program_name = link_tag.get_text(strip=True)  # Program Name
                        link = link_tag.get('href')  # Program Link
                        data.append({'Program Name': program_name, 'Link': link})

    # Step 10: Create a DataFrame from the extracted data
    df = pd.DataFrame(data)

    # Step 11: Display the DataFrame
    print("Extracted Program Data:")
    print(df)

    # Optional: Save the DataFrame to a CSV file
    df.to_csv('programs.csv', index=False)

else:
    print(f"Failed to retrieve the webpage. Status Code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Define a function to extract data from a single page
def extract_data_from_page(url):
    response = requests.get(url)
    data = []

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the table with class 'styled-table'
        table = soup.find('table', {'class': 'styled-table'})

        if table:
            tbody = table.find('tbody')
            rows = tbody.find_all('tr')

            for row in rows:
                columns = row.find_all('td')
                for column in columns:
                    list_items = column.find_all('li')
                    for item in list_items:
                        link_tag = item.find('a')
                        if link_tag:
                            program_name = link_tag.get_text(strip=True)  # Program Name
                            link = link_tag.get('href')  # Program Link
                            data.append({'Program Name': program_name, 'Link': link})

    return data

# Step 2: Loop through pages 1 to 13 and extract data from each page
base_url = 'https://www.mnsu.edu/academics/degree-programs-and-certificates/?p={}&q=&SelectedCategoryMobile=0&category=0&selectedAlpha=&selectedAlphaMobile='
all_data = []

for page in range(0, 14):  # Looping from page 1 to 13
    url = base_url.format(page)
    #print(f"Extracting data from: {url}")
    page_data = extract_data_from_page(url)
    print(page_data)
    all_data.extend(page_data)  # Combine data from all pages

# Step 3: Create a DataFrame from the extracted data
df = pd.DataFrame(all_data)

# Step 4: Display the DataFrame
print("Extracted Program Data from All Pages:")
#print(df)

# Optional: Save the DataFrame to a CSV file
df.to_csv('all_programs.csv', index=False)
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup

# Step 1: Configure Selenium WebDriver (using Chrome)
service = Service('path/to/chromedriver')  # Replace with your ChromeDriver path
driver = webdriver.Chrome()

# Step 2: List of URLs to scrape
urls = [
    'https://cob.mnsu.edu/academics/accounting/master-of-accounting/'
    # Add more URLs as needed
]

# Step 3: Function to extract content from a single URL
def extract_content_from_url(url):
    driver.get(url)  # Open the URL using Selenium
    time.sleep(10)  # Wait 10 seconds to ensure the page fully loads

    # Parse the loaded page with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Find the main content div with class 'container-fluid'
    main_div = soup.find('div', {'class': 'container-fluid'})

    if main_div:
        # Remove any header or footer elements inside the main div
        for tag in main_div.find_all(['header', 'footer']):
            tag.decompose()  # Remove these tags and their content

        # Extract the text from the cleaned main div
        text = main_div.get_text(separator='\n', strip=True)

        # Generate a dynamic filename based on the URL
        filename = url.split('/')[-2] + ".txt"

        # Save the extracted text to a file
        with open(filename, "w", encoding="utf-8") as file:
            file.write(text)

        print(f"Content from {url} saved to {filename}")
    else:
        print(f"Main div with class 'container-fluid' not found in {url}")

# Step 4: Loop through URLs and extract content from each
for url in urls:
    extract_content_from_url(url)

# Step 5: Close the Selenium WebDriver
driver.quit()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

# Create a directory to store the text files
os.makedirs("Minnesota_State_University_1/", exist_ok=True)

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    degree_name = row['Program Name']
    url = row['Link']

    # Send a GET request to fetch the HTML content of the webpage
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

         # Find the div with specific id and class
        body_container =  soup.find('div', {'class': 'container-fluid'})

        if body_container:
            # Remove any header or footer elements inside the main div (if they exist)
            for tag in body_container.find_all(['header', 'footer']):
                tag.decompose()  # Remove these tags and their content
            # Extract all text from the main content
            text_content = body_container.get_text(separator='\n', strip=True)

            # Create a valid fil
            #ename by replacing invalid characters
            filename = f"Minnesota_State_University_1/{degree_name.replace('/', '_').replace(' ', '_')}.txt"

            # Save the link and text content to the file
            with open(filename, "w", encoding="utf-8") as file:
                # Write the link at the top
                file.write(f"Link: {url}\n\n")
                # Write the extracted content below the link
                file.write(text_content)

            print(f"Content successfully saved to '{filename}'.")
        else:
            print(f"Main content not found for {degree_name}.")
    else:
        print(f"Failed to retrieve the webpage for {degree_name}. Status code: {response.status_code}")
```
