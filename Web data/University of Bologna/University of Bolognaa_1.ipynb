```python
from bs4 import BeautifulSoup

# Open and read the local HTML file
with open('page.html', 'r', encoding='utf-8') as file:
    content = file.read()

# Parse the content with BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')

# Find all containers with the 'items-container' class
containers = soup.find_all('div', class_='items-container')

program_data = []

# Loop through each container and extract data
for container in containers:
    items = container.find_all('div', class_='item')
    for item in items:
        # Extract the program name
        program_name_element = item.find('h3')
        program_name = program_name_element.text.strip() if program_name_element else None

        # Extract the program link
        link_element = item.find('a', class_='umtrack')
        program_link = link_element['href'] if link_element else None

        # Append the program name and link to the list
        if program_name and program_link:
            program_data.append({"Program Name": program_name, "Link": program_link})

# Print the extracted data
for program in program_data:
    print(f"Program Name: {program['Program Name']}")
    print(f"Link: {program['Link']}\n")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

# Create a directory to store the text files
os.makedirs("Milan_university/", exist_ok=True)

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    degree_name = row['Program Name']
    url = row['Link']

    # Send a GET request to fetch the HTML content of the webpage
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

         # Find the div with specific id and class
        body_container =  soup.find("main", class_="content-wrapper")

        if body_container:
            # Remove any header or footer elements inside the main div (if they exist)
            for tag in body_container.find_all(['header', 'footer']):
                tag.decompose()  # Remove these tags and their content
            # Extract all text from the main content
            text_content = body_container.get_text(separator='\n', strip=True)

            # Create a valid filename by replacing invalid characters
            filename = f"Milan_university/{degree_name.replace('/', '_').replace(' ', '_').replace('|','_')}.txt"

            # Save the link and text content to the file
            with open(filename, "w", encoding="utf-8") as file:
                # Write the link at the top
                file.write(f"Link: {url}\n\n")
                # Write the extracted content below the link
                file.write(text_content)

            print(f"Content successfully saved to '{filename}'.")
        else:
            print(f"Main content not found for {degree_name}.")
    else:
        print(f"Failed to retrieve the webpage for {degree_name}. Status code: {response.status_code}")
```
