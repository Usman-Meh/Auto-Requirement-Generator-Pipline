```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage
url = "https://www.unitn.it/en/ateneo/1819/programmes-of-study"

# Send a GET request to the webpage
response = requests.get(url)
response.raise_for_status()  # Ensure the request was successful

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")

# Initialize a list to store program data
programs = []

# Locate all 'h3' tags (program names) and corresponding 'a' tags (links)
for program in soup.find_all("h3"):
    # Get the program name
    name = program.get_text(strip=True)
    
    # Find the first link within the sibling <p> tag
    link_tag = program.find_next_sibling("p").find("a", href=True)
    if link_tag:
        link = link_tag["href"]  # Extract the link
    else:
        link = None  # Handle cases where no link is present
    
    # Append the data to the list
    programs.append({"Program Name": name, "Link": link})

# Convert the data to a Pandas DataFrame
df = pd.DataFrame(programs)

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# URL of the webpage
url = "https://www.unitn.it/en/ateneo/1819/programmes-of-study#cdlmcu"

# Initialize the Selenium WebDriver (e.g., for Chrome)
driver = webdriver.Chrome()  # Ensure chromedriver is in PATH

# Open the webpage
driver.get(url)

# Wait for the page to load (adjust time as needed)
time.sleep(3)

# Initialize a list to store program data
programs = []

# Locate all <h3> tags (program names)
program_elements = driver.find_elements(By.TAG_NAME, "h3")

for program_element in program_elements:
    # Get the program name
    name = program_element.text

    # Find the sibling <p> element for the link
    sibling_paragraph = program_element.find_element(By.XPATH, "following-sibling::p")
    link_element = sibling_paragraph.find_element(By.TAG_NAME, "a")
    
    # Extract the link
    link = link_element.get_attribute("href") if link_element else None

    # Append the data to the list
    programs.append({"Program Name": name, "Link": link})

# Convert the list to a DataFrame
df = pd.DataFrame(programs)

# Save the DataFrame to a CSV file
df.to_csv("unitn_programs_selenium.csv", index=False)

# Print the DataFrame
print(df)

# Close the browser
driver.quit()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from googletrans import Translator
import os

# Create a directory to store the text files
os.makedirs("University_of_Trento/", exist_ok=True)


# Initialize the Translator
translator = Translator()

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    degree_name = row['Program Name']
    url = row['Link']

    driver = webdriver.Chrome()
    driver.get(url)
    # Send a GET request to fetch the HTML content of the webpage
    response = requests.get(url)

    if response.status_code == 200:

        # Wait for the page to load
        time.sleep(5)  # Adjust as necessary for slower connections

        # Step 3: Get the page source
        page_source = driver.page_source

        # Step 4: Parse with BeautifulSoup
        soup = BeautifulSoup(page_source, 'html.parser')

        body = soup.find('div', class_='page')

        if body:
            # Exclude header and footer
            header = body.find('header')
            footer = body.find('footer')
            if header:
                header.extract()
            if footer:
                footer.extract()

            # Extract all text from the modified body tag
            all_text = body.get_text(separator='\n', strip=True)
            
            # Translate the text to English
            translated_text = translator.translate(all_text, src='it', dest='en').text
            
            # Create a valid filename by replacing invalid characters
            filename = f"University_of_Trento/{degree_name.replace('/', '_').replace(' ', '_').replace('|','_')}.txt"

            # Save the link and text content to the file
            with open(filename, "w", encoding="utf-8") as file:
                # Write the link at the top
                file.write(f"Link: {url}\n\n")
                # Write the extracted content below the link
                file.write(all_text)

            print(f"Content successfully saved to '{filename}'.")
        else:
            print(f"Main content not found for {degree_name}.")
    else:
        print(f"Failed to retrieve the webpage for {degree_name}. Status code: {response.status_code}")
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from googletrans import Translator
import os
import time

# Create a directory to store the text files
os.makedirs("University_of_Trento/", exist_ok=True)

# Initialize the Translator
translator = Translator()

# Initialize the WebDriver (Make sure ChromeDriver is in PATH)
driver = webdriver.Chrome()

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    degree_name = row['Program Name']
    url = row['Link']

    try:
        driver.get(url)
        time.sleep(5)  # Wait for the page to load
        page_source = driver.page_source

        # Parse the page source with BeautifulSoup
        soup = BeautifulSoup(page_source, 'html.parser')
        body = soup.find('div', class_='page')

        if body:
            # Exclude header and footer
            header = body.find('header')
            footer = body.find('footer')
            if header:
                header.extract()
            if footer:
                footer.extract()

            # Extract all text from the modified body tag
            all_text = body.get_text(separator='\n', strip=True)
            
            # Translate the text to English
            translated_text = translator.translate(all_text, src='it', dest='en').text
            
            # Create a valid filename by replacing invalid characters
            filename = f"University_of_Trento/{degree_name.replace('/', '_').replace(' ', '_').replace('|','_')}.txt"

            # Save the link and text content to the file
            with open(filename, "w", encoding="utf-8") as file:
                # Write the link at the top
                file.write(f"Link: {url}\n\n")
                # Write the translated content below the link
                file.write(translated_text)

            print(f"Content successfully saved to '{filename}'.")
        else:
            print(f"Main content not found for {degree_name}.")
    except Exception as e:
        print(f"Error processing {degree_name}: {e}")

# Close the WebDriver
driver.quit()
```
