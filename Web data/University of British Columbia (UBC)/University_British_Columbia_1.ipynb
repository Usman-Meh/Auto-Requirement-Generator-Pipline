```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# URL of the page to scrape
url = 'https://www.grad.ubc.ca/prospective-students/graduate-degree-programs'

# Set up Selenium WebDriver (ensure chromedriver is installed and in PATH)
driver = webdriver.Chrome()

# Open the webpage with Selenium
driver.get(url)

# Allow the page to load initially
time.sleep(40)

# Wait for the page content to load
WebDriverWait(driver, 20).until(
    EC.presence_of_element_located((By.CLASS_NAME, 'views-field-title'))
)

# Get the page source and parse with BeautifulSoup
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')

# Initialize lists to store program data
all_programs = []
specializations_ = []
programs_ = []
links_ = []

# Find all table rows with program details
rows = soup.find_all('tr', class_=['odd', 'even'])

# Extract program information from each row
for row in rows:
    specialization = row.find(
        'td', class_='views-field-field-prog-specialization').get_text(strip=True)
    title_td = row.find('td', class_='views-field-title')
    title = title_td.get_text(strip=True)
    link = title_td.find('a')['href']
    full_link = f"https://www.grad.ubc.ca{link}"

    # Store the program information
    all_programs.append({
        'Specialization': specialization,
        'Program Title': title,
        'Link': full_link
    })

    # Collect individual program details into lists
    specializations_.append(specialization)
    programs_.append(title)
    links_.append(full_link)

# Close the driver
driver.quit()

# Print all extracted programs
for program in all_programs:
    print(f"Specialization: {program['Specialization']}")
    print(f"Program Title: {program['Program Title']}")
    print(f"Link: {program['Link']}\n")
```
