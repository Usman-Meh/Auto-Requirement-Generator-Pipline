```python
import requests
from bs4 import BeautifulSoup
import re

# URL to scrape
url = "https://apply.unifi.it/courses/search/id/900481-owjor8"

# Send a GET request to fetch the page content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Select the div elements containing course information
courses = soup.select('div.ui.raised.result.segment')

# Extract program names and links
programs = []
base_url = "https://apply.unifi.it"  # Base URL for relative links

for course in courses:
    # Find program name
    program_name_tag = course.find('a', aria_label=True)
    if program_name_tag:
        program_name = program_name_tag.text.strip()
        
        # Find link and complete URL
        program_link = program_name_tag['href']
        full_link = f"{base_url}{program_link}"
        
        # Append the extracted data to the programs list
        programs.append({'Program Name': program_name, 'Link': full_link})

# Output the extracted data
for program in programs:
    print(f"Program Name: {program['Program Name']}, Link: {program['Link']}")
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

# Set up Selenium WebDriver
driver = webdriver.Chrome()  # Ensure you have ChromeDriver installed
driver.get("https://www.unifi.it/en/study-us/degree-programmes/second-cycle-degrees")

# Wait for the page to fully load
time.sleep(5)  # Adjust sleep time as needed

# Get page source and parse with BeautifulSoup
soup = BeautifulSoup(driver.page_source, "html.parser")

# Extract program names and links
programs = []
for item in soup.select(".ui.raised.result.segment .seven.wide.column a"):
    name = item.get_text(strip=True)
    link = item.get("href")
    full_link = f"https://apply.unifi.it{link}" if link else None
    programs.append((name, full_link))

# Print results
for program in programs:
    print(f"Program Name: {program[0]}, Link: {program[1]}")

# Close the WebDriver
driver.quit()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

# Create a directory to store the text files
os.makedirs("university_florence/", exist_ok=True)

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    degree_name = row['Name']
    url = row['Link']
    degree_name = clean_degree_name(degree_name)
    

    # Send a GET request to fetch the HTML content of the webpage
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

         # Find the div with specific id and class
        body_container =  soup.find("div", class_="pusher")

        if body_container:
            # Remove any header or footer elements inside the main div (if they exist)
            for tag in body_container.find_all(['header', 'footer']):
                tag.decompose()  # Remove these tags and their content
            # Extract all text from the main content
            text_content = body_container.get_text(separator='\n', strip=True)

            # Create a valid filename by replacing invalid characters
            filename = f"university_florence/{degree_name.replace('/', '_').replace(' ', '_').replace('|','_')}.txt"

            # Save the link and text content to the file
            with open(filename, "w", encoding="utf-8") as file:
                # Write the link at the top
                file.write(f"Link: {url}\n\n")
                # Write the extracted content below the link
                file.write(text_content)

            print(f"Content successfully saved to '{filename}'.")
        else:
            print(f"Main content not found for {degree_name}.")
    else:
        print(f"Failed to retrieve the webpage for {degree_name}. Status code: {response.status_code}")
```

```python
import re  # Import regex library for cleanup
# Function to clean the degree name for use in a filename
def clean_degree_name(degree_name):
    # Remove any text within square brackets and replace invalid characters
    cleaned_name = re.sub(r'\[.*?\]', '', degree_name)  # Remove text within square brackets
    cleaned_name = re.sub(r'[^a-zA-Z0-9_\- ]', '', cleaned_name)  # Keep alphanumeric, spaces, underscores, and hyphens
    return cleaned_name.strip().replace(' ', '_')  # Replace spaces with underscores for filename
```
