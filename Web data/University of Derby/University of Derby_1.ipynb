```python
import requests
from bs4 import BeautifulSoup

# URL of the page
url = "https://www.derby.ac.uk/postgraduate/courses/"

# Send a GET request to fetch the HTML content of the page
response = requests.get(url)
# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find all course items based on the HTML structure
courses = soup.find_all("div", class_="landing-page-links-item")

# Extract and print the program names and links
for course in courses:
    # Extract the link
    link_tag = course.find("a", class_="landing-page-links-item-a")
    link = "https://www.derby.ac.uk" + link_tag.get("href") if link_tag else None

    # Extract the program name
    name_tag = course.find("div", class_="landing-page-links-item-title")
    name = name_tag.get_text(strip=True) if name_tag else None

    # Print or save the result
    if name and link:
        print(f"Program Name: {name}")
        print(f"Link: {link}")
        print("-" * 40)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the page
url = "https://www.derby.ac.uk/courses/a-z/"

# Send a GET request to fetch the HTML content of the page
response = requests.get(url)
# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# List to store program details
program_data = []

# Find all course items based on the HTML structure
courses = soup.find_all("div", class_="landing-page-links-item")

# Extract program names and links and store them in the list
for course in courses:
    # Extract the link
    link_tag = course.find("a", class_="landing-page-links-item-a")
    link = "https://www.derby.ac.uk" + link_tag.get("href") if link_tag else None

    # Extract the program name
    name_tag = course.find("div", class_="landing-page-links-item-title")
    name = name_tag.get_text(strip=True) if name_tag else None

    # Add to the list if both name and link are present
    if name and link:
        program_data.append({"Program Name": name, "Link": link})

# Create a DataFrame
df = pd.DataFrame(program_data)

# Display the DataFrame
print(df)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

# Create a directory to store the text files
os.makedirs("University_of_Derby/", exist_ok=True)

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    degree_name = row['Program Name']
    url = row['Link']

    # Send a GET request to fetch the HTML content of the webpage
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        for script_or_style in soup(['script', 'style']):
            script_or_style.decompose()  # Remove these elements

        if soup:
            # Remove any header or footer elements inside the main div (if they exist)
            for tag in soup.find_all(['header', 'footer']):
                tag.decompose()  # Remove these tags and their content
            # Extract all text from the main content
            text_content = soup.get_text(separator='\n', strip=True)

            # Create a valid filename by replacing invalid characters
            filename = f"University_of_Derby/{degree_name.replace('/', '_').replace(' ', '_')}.txt"

            # Save the link and text content to the file
            with open(filename, "w", encoding="utf-8") as file:
                # Write the link at the top
                file.write(f"Link: {url}\n\n")
                # Write the extracted content below the link
                file.write(text_content)

            print(f"Content successfully saved to '{filename}'.")
        else:
            print(f"Main content not found for {degree_name}.")
    else:
        print(f"Failed to retrieve the webpage for {degree_name}. Status code: {response.status_code}")
```
