```python
import requests
from bs4 import BeautifulSoup

# URL of the page you want to scrape
url = 'https://degrees.wsu.edu/'


# Set up headers with a User-Agent
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36'
}

# Send a request to the website
response = requests.get(url,headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the content of the page
    soup = BeautifulSoup(response.content, 'html.parser')
    print('yes')
    #print(soup)


    # Find the main container with the specified class
    program_cards = soup.find_all('div', class_='wsu-degree-finder__degrees-container ')
    print(program_cards)

    # Initialize a list to store program details
    programs = []

    # Loop through the found program cards
    for card in program_cards:
        # Find all individual program cards
        individual_cards = card.find_all('div', class_='wsu-card wsu-degree-profile__related-card wsu-card--style-horizontal-25')
        
        for individual_card in individual_cards:
            # Extract program name and link
            header = individual_card.find('div', class_='wsu-card__content').find('div', class_='wsu-card__header')
            title = header.find('h2', class_='wsu-title')
            
            if title and title.a:
                program_name = title.a.text.strip()  # Program name
                program_link = title.a['href']       # Program link
                
                # Store the result
                programs.append({'name': program_name, 'link': program_link})

    # Print the extracted program names and links
    for program in programs:
        print(f"Program Name: {program['name']}, Program Link: {program['link']}")

else:
    print(f"Failed to retrieve content. Status code: {response.status_code}")
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time

# Set up Selenium WebDriver (make sure to replace the path with your chromedriver location)
driver_path = 'path_to_your_chromedriver'
driver = webdriver.Chrome()

# URL of the page to scrape
url = "https://degrees.wsu.edu/"
program_names = []
program_links = []

# Open the page using Selenium
driver.get(url)

# Wait for the JavaScript to load the content (you may need to adjust the sleep time)
time.sleep(10)

# Get the page source after JavaScript has rendered the content
page_source = driver.page_source

# Use BeautifulSoup to parse the dynamically loaded content
soup = BeautifulSoup(page_source, 'html.parser')

# Find all program sections
program_sections = soup.find_all('h2', class_='wsu-title')

# Extracting program names and links
for section in program_sections:
    program_name_tag = section.find('a')
    if program_name_tag:
        program_name = program_name_tag.get_text(strip=True)
        program_link = program_name_tag['href']
        
        print(f"Program Name: {program_name}")
        print(f"Program Link: {program_link}\n")
        
        program_names.append(program_name)
        program_links.append(program_link)

# Optionally, convert the program links to a DataFrame and save to CSV
programs_df = pd.DataFrame({
    'Program Name': program_names,
    'Program Link': program_links
})
programs_df.to_csv('W_program_links.csv', index=False)

# Close the Selenium WebDriver
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage to scrape
url = "https://degrees.wsu.edu/degree/english/"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36'
}


# Send a GET request to fetch the HTML content of the webpage
response = requests.get(url,headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the main content section by its ID or class (as per the provided sample)
    main_content = soup.find('main', id='wsu-content')

    # Extract all the text from the main content
    if main_content:
        text_content = main_content.get_text(separator='\n', strip=True)

        # Save the text content to a file
        with open("First_file.txt", "w", encoding="utf-8") as file:
            file.write(text_content)

        print("Content successfully saved to 'extracted_content.txt'.")
    else:
        print("Main content not found.")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

# Create a directory to store the text files
os.makedirs("Washington_state_university/", exist_ok=True)

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36'
}

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    degree_name = row['Program Name']
    url = row['Program Link']

    # Send a GET request to fetch the HTML content of the webpage
    response = requests.get(url,headers=headers)

    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

         # Find the div with specific id and class
        body_container =  soup.find('main', id='wsu-content')

        if body_container:
            # Remove any header or footer elements inside the main div (if they exist)
            for tag in body_container.find_all(['header', 'footer']):
                tag.decompose()  # Remove these tags and their content
            # Extract all text from the main content
            text_content = body_container.get_text(separator='\n', strip=True)

            # Create a valid fil
            #ename by replacing invalid characters
            filename = f"Washington_state_university/{degree_name.replace('/', '_').replace(' ', '_')}.txt"

            # Save the link and text content to the file
            with open(filename, "w", encoding="utf-8") as file:
                # Write the link at the top
                file.write(f"Link: {url}\n\n")
                # Write the extracted content below the link
                file.write(text_content)

            print(f"Content successfully saved to '{filename}'.")
        else:
            print(f"Main content not found for {degree_name}.")
    else:
        print(f"Failed to retrieve the webpage for {degree_name}. Status code: {response.status_code}")
```
