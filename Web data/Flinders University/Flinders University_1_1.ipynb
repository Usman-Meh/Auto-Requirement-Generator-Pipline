```python
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = "https://www.flinders.edu.au/international/business"

# Sending a request to fetch the content of the webpage
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Find the section containing the course list
course_list = soup.find("ul", class_="course_list")

# Loop through each course item and extract the program name and link
for course in course_list.find_all("a", class_="course_list_item course_link"):
    program_name = course.find("span", class_="course_name_label").get_text(strip=True)
    program_link = course["href"]
    print(f"Program Name: {program_name}")
    print(f"Program Link: {program_link}")
    print("-" * 50)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the page to scrape
url = "https://www.flinders.edu.au/international/Social-work"

# Sending a request to fetch the content of the webpage
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# List to store program data
program_data = []

# Find all sections containing course lists
accordion_items = soup.find_all("div", class_="accordion_item")

# Loop through each accordion section
for section in accordion_items:
    # Get the section heading (e.g., "Undergraduate", "Postgraduate")
    section_heading = section.find("a", class_="item_heading").get_text(strip=True)
    
    # Find the course list within each section
    course_list = section.find("ul", class_="course_list")
    
    # If course list exists, extract program names and links
    if course_list:
        for course in course_list.find_all("a", class_="course_list_item course_link"):
            program_name = course.find("span", class_="course_name_label").get_text(strip=True)
            program_link = course["href"]
            # Append program information as a dictionary to the list
            program_data.append({
                "Section": section_heading,
                "Program Name": program_name,
                "Program Link": program_link
            })

# Create a DataFrame from the list
df = pd.DataFrame(program_data)

```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

# Create a directory to store the text files
os.makedirs("Flinders_University/", exist_ok=True)

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    degree_name = row['Program Name']
    url = row['Program Link']

    # Send a GET request to fetch the HTML content of the webpage
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

         # Find the div with specific id and class
        body_container =  soup.find("div", class_="body_sub_page")

        if body_container:
            # Remove any header or footer elements inside the main div (if they exist)
            for tag in body_container.find_all(['header', 'footer']):
                tag.decompose()  # Remove these tags and their content
            # Extract all text from the main content
            text_content = body_container.get_text(separator='\n', strip=True)

            # Create a valid filename by replacing invalid characters
            filename = f"Flinders_University/{degree_name.replace('/', '_').replace(' ', '_').replace('|','_')}.txt"

            # Save the link and text content to the file
            with open(filename, "w", encoding="utf-8") as file:
                # Write the link at the top
                file.write(f"Link: {url}\n\n")
                # Write the extracted content below the link
                file.write(text_content)

            print(f"Content successfully saved to '{filename}'.")
        else:
            print(f"Main content not found for {degree_name}.")
    else:
        print(f"Failed to retrieve the webpage for {degree_name}. Status code: {response.status_code}")
```
