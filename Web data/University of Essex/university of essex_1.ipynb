```python
import requests
from bs4 import BeautifulSoup

# URL to scrape
url = "https://www.essex.ac.uk/course-search?collection=uoe-courses-meta&start_rank=1"

# Make a GET request to fetch the HTML content
response = requests.get(url)
if response.status_code == 200:
    # Parse the HTML using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all program names and links
    courses = soup.find_all('a', class_='course-search-card course-search-card--labeled')
    
    # Extract and display the program names and links
    for course in courses:
        program_name = course.find('h3', class_='course-search-card__title purple').get_text(strip=True)
        link = course['href']
        print(f"Program Name: {program_name}")
        print(f"Link: {link}")
        print("-" * 50)
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL
base_url = "https://www.essex.ac.uk/course-search?collection=uoe-courses-meta&start_rank="

# List to store extracted data
data = []

# Iterate through pages (increment by 10, from 1 to 841)
for page_number in range(1, 842, 10):
    # Construct the URL with the current page number
    url = f"{base_url}{page_number}"
    print(f"Fetching data from: {url}")
    
    # Make a GET request to fetch the HTML content
    response = requests.get(url)
    
    if response.status_code == 200:
        # Parse the HTML using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find all program entries
        courses = soup.find_all('a', class_='course-search-card course-search-card--labeled')
        
        # Extract program name and link for each course
        for course in courses:
            program_name = course.find('h3', class_='course-search-card__title purple').get_text(strip=True)
            link = course['href']
            
            # Append to the data list
            data.append({"Program Name": program_name, "Link": link})
    else:
        print(f"Failed to fetch data from page {page_number}. Status code: {response.status_code}")

# Save the data into a pandas DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)


```

```python
# Save to a CSV file
df.to_csv("essex_links.csv", index=False)
print("Data has been saved to 'programs_and_links.csv'")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

# Create a directory to store the text files
os.makedirs("University_of_Essex_1/", exist_ok=True)

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    degree_name = row['Program Name']
    url = row['Link']

    # Send a GET request to fetch the HTML content of the webpage
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

         # Find the div with specific id and class
        body_container =  soup.find('div', {'class': 'wrapper'})

        if body_container:
            # Remove any header or footer elements inside the main div (if they exist)
            for tag in body_container.find_all(['header', 'footer']):
                tag.decompose()  # Remove these tags and their content
            # Extract all text from the main content
            text_content = body_container.get_text(separator='\n', strip=True)

            # Create a valid filename by replacing invalid characters
            filename = f"University_of_Essex_1/{degree_name.replace('/', '_').replace(' ', '_')}.txt"

            # Save the link and text content to the file
            with open(filename, "w", encoding="utf-8") as file:
                # Write the link at the top
                file.write(f"Link: {url}\n\n")
                # Write the extracted content below the link
                file.write(text_content)

            print(f"Content successfully saved to '{filename}'.")
        else:
            print(f"Main content not found for {degree_name}.")
    else:
        print(f"Failed to retrieve the webpage for {degree_name}. Status code: {response.status_code}")
```
