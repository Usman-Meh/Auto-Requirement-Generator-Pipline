```python
from bs4 import BeautifulSoup

# Load the HTML file
with open('page.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Find all program names and links
programs = []
for leaf in soup.select('.corsi-leaf a'):
    program_name = leaf.get_text(strip=True)
    program_link = 'https://unipi.coursecatalogue.cineca.it' + leaf['href']
    programs.append({'name': program_name, 'link': program_link})

# Display the extracted programs
for program in programs:
    print(f"Program Name: {program['name']}")
    print(f"Link: {program['link']}\n")
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import os
import time

# Replace 'path_to_chromedriver' with the actual path to your WebDriver
service = Service('path_to_chromedriver')
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run in headless mode to avoid opening the browser window
driver = webdriver.Chrome()

# Create a directory to store the text files
os.makedirs("Pisa_university/", exist_ok=True)

# Assuming `df` is your DataFrame with 'Name' and 'Link' columns
for index, row in df.iterrows():
    degree_name = row['name']
    url = row['link']
    degree_name = clean_degree_name(degree_name)
    # Load the page using Selenium
    driver.get(url)
    
    try:
        # Wait for the main content to load
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'mat-typography'))
        )
        time.sleep(2)  # Optional: wait for a few more seconds if needed

        # Get the page source and parse with BeautifulSoup
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')

        # Find the main body container
        body_container = soup.find("body", class_="mat-typography")
        
        if body_container:
            # Remove any header or footer elements inside the main div
            for tag in body_container.find_all(['header', 'footer']):
                tag.decompose()

            # Extract all text from the main content
            text_content = body_container.get_text(separator='\n', strip=True)

            # Clean degree name for valid filename
            filename = f"Pisa_university/{degree_name.replace('/', '_').replace(' ', '_').replace('|','_')}.txt"

            # Save the link and text content to the file
            with open(filename, "w", encoding="utf-8") as file:
                # Write the link at the top
                file.write(f"Link: {url}\n\n")
                # Write the extracted content below the link
                file.write(text_content)

            print(f"Content successfully saved to '{filename}'.")
        else:
            print(f"Main content not found for {degree_name}.")
    except Exception as e:
        print(f"Error processing {degree_name}: {e}")

# Close the Selenium WebDriver
driver.quit()
```

```python
import re  # Import regex library for cleanup
# Function to clean the degree name for use in a filename
def clean_degree_name(degree_name):
    # Remove any text within square brackets and replace invalid characters
    cleaned_name = re.sub(r'\[.*?\]', '', degree_name)  # Remove text within square brackets
    cleaned_name = re.sub(r'[^a-zA-Z0-9_\- ]', '', cleaned_name)  # Keep alphanumeric, spaces, underscores, and hyphens
    return cleaned_name.strip().replace(' ', '_')  # Replace spaces with underscores for filename
```
