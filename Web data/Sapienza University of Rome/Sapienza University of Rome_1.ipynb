```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage you want to scrape
url = "https://corsidilaurea.uniroma1.it/en/solr-offerta-formativa-t?search_api_views_fulltext="

# Make a request to get the page content
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table containing the program data
table = soup.find('table', class_='views-table')

# Initialize lists to store program names and URLs
programs = []

# Iterate through each row in the table body
for row in table.find('tbody').find_all('tr'):
    # Find the link within the 'Programmes' column
    program_link = row.find('td', class_='views-field-title-field').find('a')
    
    # Get the program name and URL
    program_name = program_link.get_text(strip=True)
    program_url = program_link['href']
    
    # Store the program name and full URL
    programs.append({
        'name': program_name,
        'url': program_url
    })

# Display the extracted data
for program in programs:
    print(f"Program Name: {program['name']}")
    print(f"URL: {program['url']}\n")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL of the webpage
base_url = "https://corsidilaurea.uniroma1.it/en/solr-offerta-formativa-t?search_api_views_fulltext="

# Initialize a list to store program data
programs = []

# Loop through each page from 0 to 7
for page in range(8):
    # Construct the URL for each page
    url = f"{base_url}&page={page}"
    
    # Make a request to get the page content
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the table containing the program data
    table = soup.find('table', class_='views-table')
    
    # Check if the table exists on the page
    if table:
        # Iterate through each row in the table body
        for row in table.find('tbody').find_all('tr'):
            # Find the link within the 'Programmes' column
            program_link = row.find('td', class_='views-field-title-field').find('a')
            
            # Get the program name and URL
            program_name = program_link.get_text(strip=True)
            program_url = program_link['href']
            
            # Store the program name and full URL
            programs.append({
                'Program Name': program_name,
                'Program URL': program_url
            })
    else:
        print(f"No table found on page {page}")

# Convert the data to a DataFrame for easy viewing
df = pd.DataFrame(programs)

# Display the extracted data
print(df)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os

# Create a directory to store the text files
os.makedirs("Sapienza_university_1/", exist_ok=True)

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    degree_name = row['name']
    url = row['url']
    
    degree_name = clean_degree_name(degree_name)

    # Send a GET request to fetch the HTML content of the webpage
    response = requests.get(url)

    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

         # Find the div with specific id and class
        body_container =  soup.find("section", class_="clearfix section")

        if body_container:
            # Remove any header or footer elements inside the main div (if they exist)
            for tag in body_container.find_all(['header', 'footer']):
                tag.decompose()  # Remove these tags and their content
            # Extract all text from the main content
            text_content = body_container.get_text(separator='\n', strip=True)

            # Create a valid filename by replacing invalid characters
            filename = f"Sapienza_university_1/{degree_name.replace('/', '_').replace(' ', '_').replace('|','_')}.txt"

            # Save the link and text content to the file
            with open(filename, "w", encoding="utf-8") as file:
                # Write the link at the top
                file.write(f"Link: {url}\n\n")
                # Write the extracted content below the link
                file.write(text_content)

            print(f"Content successfully saved to '{filename}'.")
        else:
            print(f"Main content not found for {degree_name}.")
    else:
        print(f"Failed to retrieve the webpage for {degree_name}. Status code: {response.status_code}")
```

```python
import re  # Import regex library for cleanup
# Function to clean the degree name for use in a filename
def clean_degree_name(degree_name):
    # Remove any text within square brackets and replace invalid characters
    cleaned_name = re.sub(r'\[.*?\]', '', degree_name)  # Remove text within square brackets
    cleaned_name = re.sub(r'[^a-zA-Z0-9_\- ]', '', cleaned_name)  # Keep alphanumeric, spaces, underscores, and hyphens
    return cleaned_name.strip().replace(' ', '_')  # Replace spaces with underscores for filename
```
